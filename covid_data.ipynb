```python
# Get API_KEY from https://covidactnow.org/data-api
import pandas as pd
import config
import requests
from config import API_KEY
```

```python
# Time series for all states
covid_df = pd.read_csv(f'https://api.covidactnow.org/v2/states.timeseries.csv?apiKey={API_KEY}')
```

```python
#Drop columns that are irrelevant
covid_df_clean1 = covid_df.drop(columns=['country','county','fips','lat','long','actuals.contactTracers', 'actuals.hospitalBeds.capacity','actuals.hospitalBeds.currentUsageTotal','unused1','actuals.icuBeds.capacity','actuals.icuBeds.currentUsageTotal','unused3', 'unused4','metrics.icuCapacityRatio','locationId','actuals.vaccinesDistributed','actuals.vaccinationsInitiated', 'cdcTransmissionLevel','actuals.vaccinationsAdditionalDose', 'metrics.vaccinationsAdditionalDoseRatio' ])
```

```python
#drop null 
covid_df_clean1.dropna()
```

```python
covid_df_clean1.head()
```

```python
#filter the dataframe
covid_df_filtered = covid_df_clean1.loc[(covid_df_clean1['date'] >= '2021-03-01') & (covid_df_clean1['date'] < '2022-02-28')]
covid_df_filtered.head(10)
```

```python
covid_df_filtered.to_csv('Resources\COVID_Filtered.csv', index = False)
```
